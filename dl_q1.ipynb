```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from keras.datasets import fashion_mnist
```

```python
import tensorflow as tf
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
## checks if the  x_train doesnot contain any of the nan , inf,  or -inf
np.isfinite(x_train).all()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
print("x_train:",x_train.shape)
print("y_train:", y_train.shape)
```


##Visualize the data

```python
import random
class_number = set()
datapoints = y_train.shape[0]
while len(class_number) != 10:
  img = random.randrange(1, 60000)
  if y_train[img] not in class_number:
    class_number.add(y_train[img])
    plt.figure()
    plt.title(class_names[y_train[img]])
    plt.imshow(x_train[img],vmin=0, vmax=255)
    plt.grid(False)
```
